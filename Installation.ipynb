# Notebook d'instalation

Dans ce notebook nous allons installer MMPose avec Cuda 12.4, avec Python en version 3.8 et en utilisant conda sous Linux.  
On commence par installer PyTorch :

```python
import subprocess
subprocess.run(["conda","install","pytorch","torchvision","-c","pytorch"])
```

Installation de openmim puis installation de mmengine et mmcv. On ne prend pas la dernière version de mmcv car elle pose problème.

```python
%pip install -U openmim
subprocess.run(["mim","install","mmengine"])
subprocess.run(["mim","install","mmcv>=2.0.0rc4,<2.2.0"])
```

Installation et build de MMPose a partir du dépot git

```python
subprocess.run(["git", "clone", "https://github.com/open-mmlab/mmpose.git"])
subprocess.run(["cd","mmpose"])
%pip install -r mmpose/requirements.txt
%pip install -v -e ./mmpose
```

Test de l'installation avec le test fourni dans la documentation :

```python
from mmpose.apis import inference_topdown, init_model
from mmpose.utils import register_all_modules

register_all_modules()

config_file = 'mmpose/td-hm_hrnet-w48_8xb32-210e_coco-256x192.py'
checkpoint_file = 'mmpose/td-hm_hrnet-w48_8xb32-210e_coco-256x192-0e67c616_20220913.pth'
model = init_model(config_file, checkpoint_file, device='cuda:0')

# please prepare an image with person
results = inference_topdown(model, 'mmpose/demo.jpg')

print(results)
```
