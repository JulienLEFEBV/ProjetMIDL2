```python
from mmpose.apis import MMPoseInferencer
import os

path="./data/videos"
videos = os.listdir(path)
videos = [f for f in videos if os.path.isfile(os.path.join(path, f))]

for video in videos:
    inferencer = MMPoseInferencer(pose3d="human3d")
    result_generator = inferencer(os.path.join(path, video),pred_out_dir='data/points',device='cuda:0')
    result = list(result_generator)

print("Done")
```
