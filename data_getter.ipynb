```python
CACHED_FOLDER_PATH = "./data/cached"
VIDEO_FOLDER_PATH = "./data/videos"

VIDEO_IDS = [
    "1f9kPJBdHl9PAW2DFCzJgPXuhWuHZfJSD"
]
VIDEO_PATH = [
    f"{VIDEO_FOLDER_PATH}/video{id}.mp4" for id in VIDEO_IDS
]
CACHED_PATH = [
    f"{CACHED_FOLDER_PATH}/{id}.npy" for id in VIDEO_IDS
]
```

```python
# creating needed files
import os

os.makedirs(CACHED_FOLDER_PATH, exist_ok=True)
os.makedirs(VIDEO_FOLDER_PATH, exist_ok=True)
```

```python
# downloading videos

import gdown

for id,video_path in zip(VIDEO_IDS, VIDEO_PATH):
    url = f"https://drive.google.com/uc?id={id}"
    gdown.download(url, video_path)
```

```python
# getting video frames 

import cv2
import numpy as np

VIDEOS = {}

for id,video_path in zip(VIDEO_IDS,VIDEO_PATH) :
    cap = cv2.VideoCapture(video_path)
    frames = []
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret : break
        frames.append(frame)

    cap.release()
    VIDEOS[id] = np.array(frames)

   
```

```python
# caching data

for id in VIDEO_IDS:
    np.save(f"./data/cached/{id}.npy", VIDEOS[id])

del VIDEOS
```
